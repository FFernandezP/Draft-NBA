
###Normalización y descripción

```python
from sklearn.preprocessing import normalize
from sklearn.cluster import KMeans
data_norm = (data - data.min())/(data.max() - data.min())
data_norm = data_norm.replace(np.nan, 0)
data_norm.describe()
```

###Clusters

```python
wcss = []            # Es la medida de la suma de las distancias de las observaciones desde sus centroides de grupo.
for i in range(1,11):
    kmeans = KMeans(n_clusters = i, max_iter= 300)
    kmeans.fit(data_norm)
    wcss.append(kmeans.inertia_)
```

##Codos Jambú

```python
plt.plot(range(1,11), wcss)
plt.title("Codo de Jambú")
plt.xlabel('Numero de clusters')
plt.ylabel('WCSS')
plt.axvline(x=3, color='r', linestyle='--')
```

Al ver el graficos a nuestro parecer el codo es en 3 clusters por eso la decision de usar ese aunque por momentos nos planteamos usar 4 pero llegamos a la conclusion que en 4 la curva deja de ser tan pronunciada

##Modelo Kmeans

```python
modelos_clust = KMeans(n_clusters = 3, max_iter= 300)
modelos_clust.fit_predict(data_norm)
```

```python
data['KMeans_Clusters'] = modelos_clust.labels_
data1
```

## Reducción dimensionalidad PCA

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=3)
pca_data_norm = pca.fit_transform(data_norm)

pca_data_norm_df = pd.DataFrame(data = pca_data_norm, columns = ['Componente_1', 'Componente_2', 'Componente_3'])

pca_nombres_data = pd.concat([pca_data_norm_df, data['KMeans_Clusters']], axis=1)
pca_nombres_data
```
